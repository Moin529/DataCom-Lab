```python
# AMI

import numpy as np
import matplotlib.pyplot as plt
```

```python
# AMI Encoding Function

def ami_encode(bits):
    encoded_signal = []
    last = -1 # Starting from -1 as initial voltage
    for bin_signal in bits:
        if bin_signal == '0':
            encoded_signal.append(0)
        elif bin_signal == '1':
            last *= -1 #Inverting polarity
            encoded_signal.append(last)
    return encoded_signal
```

```python
# AMI Decoding Function

def ami_decode(encoded_signal):
    decoded_sequence = []
    for signal in encoded_signal:
        if signal == 0:
            decoded_sequence.append('0') #low voltage
        else:
            decoded_sequence.append('1') #high voltage
    return ''.join(decoded_sequence)
```

```python
# Plotting Function

def plot(bits, encoded_signal):
    time = []
    voltage = []

    for i, level in enumerate(encoded_signal):
        # Each bit is 1 time unit wide
        time.extend([i, i + 1])
        voltage.extend([level, level])

    plt.figure(figsize=(14, 6))
    plt.plot(time, voltage, label= 'AMI Encoded Signal')
    plt.title('AMI Encoding')
    plt.xlabel('Time')
    plt.ylabel('Signal Level')
    plt.ylim(-1.5,1.5)
    plt.yticks([-1,0,1])
    plt.grid(True)
    plt.legend()

    # Bit annotations
    for i, bit in enumerate(bits):
        plt.text(i + 0.5, 1.3, bit, ha='center', va='center', fontsize=12, color='black')

    plt.show()
```

```python
#Input

binary_sequence = '10100111001'
```

```python
#Plotting Encoded Signal and Decoding Sequence

encoded_signal = ami_encode(binary_sequence)
print("Encoded Signal: ", encoded_signal)

plot(binary_sequence, encoded_signal)

decoded_sequence = ami_decode(encoded_signal)
print("Decoded Sequence: ", decoded_sequence)
```
